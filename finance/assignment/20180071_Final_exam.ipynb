# Regression vs Classification (총 40점)

- 이번 문제들에서는 Lasso example 노트북의 실현변동성(realized volatility) 데이터를 이용합니다.

- (1) (4점) 'data1.csv' 파일을 불러오고 맨 위, 맨 아래 행을 각각 5개씩 출력하세요.

```python
import pandas as pd

df = pd.read_csv('./data1.csv', index_col='Unnamed: 0')
df.head(5)
```

```python
df.tail(5)
```

- (2) (4점) 데이터 중 다우존스 지수(symbol=DJI)에 해당하는 데이터 중 'rv5' 만을 따로 뽑아내서 새로운 dataframe을 구성하고, rv5의 과거값을 lag1 부터 lag 20까지를 각각 개별적인 column으로 추가하세요.

```python
dow = df[df['Symbol']=='.DJI'][['rv5']]
dow
```

```python
for i in range(1, 21):
    dow['rv5_'+str(i)] = dow['rv5'].shift(i)
dow = dow.dropna()
dow
```

- (3) (5점) 시계열 자료가 안정적인지 검증하세요.

```python
# stationary test
from statsmodels.tsa.stattools import adfuller

print(adfuller(dow.rv5, regression='n')[1])
print(adfuller(dow.rv5, regression='c')[1])
print(adfuller(dow.rv5, regression='ct')[1])
```

어떤 regression이든 p value < 0.05이므로 귀무가설을 기각한다. 따라서 dow 지수의 rv5 자료는 stationary한 시계열 자료이다.

- (4)  (4점) Dataframe의 rv5를 y변수로, rv5를 제외한 모든 lag 변수들을 X 변수로 설정하고, sklearn의 train_test_split을 이용해 train set과 test set을 구성하세요(이때, train_test_split에서 test size는 0.3, random state는 42로 설정).

```python
from sklearn.model_selection import train_test_split
y = dow['rv5']
X = dow.drop('rv5', axis=1)
X_train_rd, X_test_rd, y_train_rd, y_test_rd = train_test_split(X, y, test_size=0.3,
                                                    random_state=42)
```

- (5) (5점) sklearn의 Ridge regression을 train set으로 훈련하고, test set에 대해 예측을 진행하십시오. Test set에 대한 LinearRegression 모형의 $R^2$ score를 도출하십시오. 위의 모형이 rv5에 대해 예측력을 가진다고 말할 수 있습니까? (단, hyperparameter는 따로 tuning하지 않고 alpha만 0.001로 설정합니다.)

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score

ridge_reg = Ridge(alpha=0.001)
ridge_reg.fit(X_train_rd, y_train_rd)
y_pred_rd = ridge_reg.predict(X_test_rd)
print('r2 score =', r2_score(y_test_rd, y_pred_rd))
```

r2 score는 1에 가까울 수록 성능이 좋고 0에 가까울 수록 성능이 나쁘다는 의미인데, 일반적으로 r square가 0.3 이상이면 종속변수에 대한 예측력을 가진다고 본다. 즉 이 모델은 rv5에 대한 예측력을 가진다고 할 수 있다.

- (6) (6점) rv5가 직전 관측치보다 상승했으면 1, 감소하거나 변하지 않았으면 0을 부여하고 이를 y변수로 설정합니다. sklearn의 Logistic regression을 train set으로 훈련하고, test set에 대한 Logistic regression 모형의 classification report를 도출하십시오(단, hyperparameter는 따로 tuning하지 않고 default 값 그대로 사용합니다). 이 결과를 바탕으로, logistic regression 모형이 rv5의 움직임에 대해 예측력을 가진다고 말할 수 있습니까?

```python
dow['target'] = 0

for i in range(1, len(dow)):
    if dow.iloc[i, 0] >  dow.iloc[i-1, 0]:
        dow.iloc[i, -1] = 1
```

```python
dow.drop('2000-02-01 00:00:00+00:00', axis=0)
```

첫번째 행은 이전의 관측치가 없으므로 삭제하였다.

```python
from sklearn.model_selection import train_test_split
y = dow['target']
X = dow.drop('target', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,
                                                    random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, classification_report


lg = LogisticRegression()
lg.fit(X_train,y_train)
predicted = lg.predict(X_test)
print(classification_report(y_test, predicted))
```

accuracy만 확인하면 정확도가 0.5니 설명력이 있다고 착각할 수 있다. 그러나 클래스별 f1-score를 보면, 클래스 1에 대해서는 0.66인데 반해 클래스 0에 대해서는 0.00이다. 이를 통해 rv5가 하락한 케이스에 대해서는 예측력이 없는 모델이라는 것을 알 수 있다. 또한 클래스 1의 세부 지표를 살펴보면, recall은 1.00인데 반해 precision은 0.50이다. 이 말은 target이 1일 때 모델이 전부 1을 예측했으나, 모델이 1을 예측했을 때 정답이 1인 경우는 절반밖에 되지 않았다는 것이다. 결국 이 모델은 대다수의 test 데이터에 대해 1을 예측하는 잘못된 모델이라는 것을 추측할 수 있다.

```python
pd.DataFrame({'pd':predicted}).value_counts()
```

실제로 예측치를 확인해보면, 전부 클래스 1뿐이다. 즉 이 모델은 rv5의 움직임에 대해서 예측력이 없다.

- (7) (6점) 다음 날의 rv5가 오늘의 rv5와 같을 것이라는 모형(즉, $\hat{rv}_{t+1} = rv_t$)의 $R^2$ score를 (5)의 test set을 이용해 도출하고, 이 결과를 바탕으로 (5)와 (6)의 결과를 해석하세요.

```python
X_test_rd = X_test_rd.sort_index()
y_test_rd = y_test_rd.sort_index()
y_predict = y_test_rd.copy()

for row in range(1, len(y_predict)):
    y_predict[row] = y_predict[row-1]
print('r2 score =', r2_score(y_test_rd, y_predict))
```

```python
y_predict
```

다음날의 rv5가 오늘의 rv5와 같으므로, 첫번째 날의 rv5가 모든 예측치가 되도록 할당해주었다. 그리고 test set과 r square score를 구해보자 음수가 나온다. 이는 종속변수의 일괄 평균값으로 예측하는 것보다 낮은 성능을 보인다는 의미이므로 엉터리 모델이라고 할 수 있다.  
이에 반해 (5)의 경우는 양수이며 30%의 설명력이 있으므로 제대로 된 모델이다. 또한 (6)의 경우는 이 엉터리 모델과 유사하게 모든 예측치가 1이었지만 accuracy는 0.5 수준으로 나왔다. 그러므로 모델링을 할 때는 단순 score만 보고 모델의 예측력을 판단하는 것이 아니라 세부적인 지표까지 참고해야 한다는 것을 알 수 있다.

- (8) (6점) (6)의 분석을 SVM과 Random Forest로 다시 진행하고(단, hyperparameter는 따로 tuning하지 않고 default 값 그대로 사용합니다), test set에 대한 각 모형의 classification report를 도출합니다.

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

전체적인 accurace score가 0.93으로 굉장히 높다. 세부적으로 살펴보면, 클래스 0과 1에 대해 모두 0.93의 높은 f1-score를 가진다. 또한 precision과 recall도 0.92~0.94 수준으로 높은 편이므로 이 모델은 rv5의 움직임에 대해 예측력이 높은 모델이라 할 수 있다.

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split, KFold, cross_val_score


svc = SVC()
svc.fit(X_train,y_train)
predicted = svc.predict(X_test)
print(classification_report(y_test, predicted))
```

전체 accuracy score가 0.81 수준으로 앞선 RandomForest보다는 낮다. 그러나 각 클래스별 f1-score가 0.79에서 0.83수준이고, precision과 recall도 약간의 차이는 있으나 크게 편향된 수준은 아니다. 그러므로 이 모델은 rv5의 움직임에 대해서 예측력이 높은 모델이라고 할 수 있다.
